```python
from PIL import Image
import torch
import os

class WasteDetDataset(torch.utils.data.Dataset):
    def __init__(self, image_dir, label_dir, transforms=None):
        self.image_dir = image_dir
        self.label_dir = label_dir
        self.transforms = transforms
        self.image_files = sorted(os.listdir(image_dir))

    def __getitem__(self, idx):
        img_path = os.path.join(self.image_dir, self.image_files[idx])
        lbl_path = os.path.join(self.label_dir,
                               os.path.splitext(self.image_files[idx])[0] + ".txt")

        img = Image.open(img_path).convert("RGB")
        w, h = img.size

        boxes = []
        labels = []

        if os.path.exists(lbl_path):
            with open(lbl_path, "r") as f:
                for line in f.readlines():
                    parts = line.strip().split()
                    if len(parts) != 5:
                        continue
                    c, cx, cy, bw, bh = map(float, parts)
                    x1 = (cx - bw / 2) * w
                    y1 = (cy - bh / 2) * h
                    x2 = (cx + bw / 2) * w
                    y2 = (cy + bh / 2) * h
                    boxes.append([x1, y1, x2, y2])
                    labels.append(int(c))

        if len(boxes) == 0:
            boxes = torch.zeros((0,4), dtype=torch.float32)
            labels = torch.zeros((0,), dtype=torch.int64)
        else:
            boxes = torch.tensor(boxes, dtype=torch.float32)
            labels = torch.tensor(labels, dtype=torch.int64)

        target = {
            "boxes": boxes,
            "labels": labels
        }

        if self.transforms:
            img = self.transforms(img)

        return img, target

    def __len__(self):
        return len(self.image_files)

print("WasteDetDataset loaded successfully! 🚀")
```

```python
import torchvision.ops as ops

def calculate_accuracy(outputs, targets, score_thresh=0.3, iou_thresh=0.5):
    """
    Accuracy = proporsi prediksi bbox yang:
      - score >= score_thresh
      - IoU >= iou_thresh
      - label pred == label GT yang dipasangkan
    """
    correct = 0
    total = 0

    for pred, tgt in zip(outputs, targets):
        gt_boxes  = tgt["boxes"]
        gt_labels = tgt["labels"]

        if gt_boxes.numel() == 0:
            continue

        pred_boxes  = pred["boxes"]
        pred_labels = pred["labels"]
        pred_scores = pred["scores"]

        keep = pred_scores >= score_thresh
        pred_boxes  = pred_boxes[keep]
        pred_labels = pred_labels[keep]

        if pred_boxes.numel() == 0:
            continue

        ious = ops.box_iou(pred_boxes, gt_boxes)   # [P, G]
        max_iou, idx = ious.max(dim=1)

        for i in range(len(max_iou)):
            total += 1
            if max_iou[i] >= iou_thresh and pred_labels[i] == gt_labels[idx[i]]:
                correct += 1

    return correct, total
```

```python
import torch.optim as optim
from torch import amp
from tqdm.auto import tqdm
import os, pickle
import pandas as pd
import torchvision.ops as ops

def calculate_accuracy(outputs, targets, score_thresh=0.3, iou_thresh=0.5):
    correct = 0
    total = 0
    for pred, tgt in zip(outputs, targets):
        gt_boxes  = tgt["boxes"]
        gt_labels = tgt["labels"]

        if gt_boxes.numel() == 0:
            continue

        pred_boxes  = pred["boxes"]
        pred_labels = pred["labels"]
        pred_scores = pred["scores"]

        keep = pred_scores >= score_thresh
        pred_boxes  = pred_boxes[keep]
        pred_labels = pred_labels[keep]

        if pred_boxes.numel() == 0:
            continue

        ious = ops.box_iou(pred_boxes, gt_boxes)
        max_iou, idx = ious.max(dim=1)

        for i in range(len(max_iou)):
            total += 1
            if max_iou[i] >= iou_thresh and pred_labels[i] == gt_labels[idx[i]]:
                correct += 1
    return correct, total


def train_ssd(model, train_loader, val_loader, device,
              num_classes_det,
              epochs=30,
              lr=1e-4,
              save_path="/content/models/ssd_mobilenet_best_from_scratch.pt",
              history_path="/content/models/ssd_mobilenet_history_from_scratch.pkl",
              csv_path="/content/models/ssd_mobilenet_results_from_scratch.csv"):

    os.makedirs(os.path.dirname(save_path), exist_ok=True)

    model.to(device)
    use_amp = (device.type == "cuda")
    scaler = amp.GradScaler("cuda", enabled=use_amp)
    optimizer = optim.AdamW(
        [p for p in model.parameters() if p.requires_grad],
        lr=lr
    )

    train_losses, val_losses = [], []
    train_accs,   val_accs   = [], []
    best_val_acc = 0.0

    for epoch in range(epochs):
        # ===== TRAIN LOSS =====
        model.train()
        running_loss = 0.0
        skipped_train = 0
        pbar = tqdm(train_loader, desc=f"[SSD] Epoch {epoch+1}/{epochs}", leave=True)
        print(f"\n==== EPOCH {epoch+1}/{epochs} ====")


        for images, targets in pbar:
            images  = [img.to(device) for img in images]
            targets = [{k: v.to(device) for k, v in t.items()} for t in targets]

            optimizer.zero_grad()

            with amp.autocast("cuda", enabled=use_amp):
                loss_out = model(images, targets)

            # kalau output bukan dict → berarti ada masalah bbox, skip batch
            if not isinstance(loss_out, dict):
                skipped_train += 1
                continue

            loss = sum(loss_out.values())

            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()

            running_loss += loss.item()
            pbar.set_postfix({"loss": f"{loss.item():.3f}"})

        efektif_batch = max(1, len(train_loader) - skipped_train)
        train_loss = running_loss / efektif_batch

        # ===== TRAIN ACCURACY =====
        model.eval()
        correct_train, total_train = 0, 0
        with torch.no_grad():
            pbar_acc = tqdm(train_loader, desc=f"[SSD] Train Acc E{epoch+1}", leave=False)
            for images, targets in pbar_acc:
                images  = [img.to(device) for img in images]
                targets = [{k: v.to(device) for k, v in t.items()} for t in targets]
                outputs = model(images)
                c, t = calculate_accuracy(outputs, targets,
                                          score_thresh=0.3, iou_thresh=0.5)
        correct_train += c
        total_train   += t

        train_acc = correct_train / max(1, total_train)

        # ===== VAL LOSS + ACC =====
        val_loss_sum = 0.0
        skipped_val = 0
        correct_val, total_val = 0, 0

        with torch.no_grad():
            for images, targets in val_loader:
                images  = [img.to(device) for img in images]
                targets = [{k: v.to(device) for k, v in t.items()} for t in targets]

                # loss
                with amp.autocast("cuda", enabled=use_amp):
                    loss_out = model(images, targets)
                if not isinstance(loss_out, dict):
                    skipped_val += 1
                    continue
                loss = sum(loss_out.values())
                val_loss_sum += loss.item()

                # accuracy
                outputs = model(images)
                c, t = calculate_accuracy(outputs, targets,
                                          score_thresh=0.3, iou_thresh=0.5)
                correct_val += c
                total_val   += t

        efektif_val_batch = max(1, len(val_loader) - skipped_val)
        val_loss = val_loss_sum / efektif_val_batch
        val_acc  = correct_val / max(1, total_val)

        train_losses.append(train_loss)
        val_losses.append(val_loss)
        train_accs.append(train_acc)
        val_accs.append(val_acc)

        print(f"[SSD] Epoch {epoch+1}/{epochs} | "
              f"Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f} | "
              f"Train Acc: {train_acc:.4f} | Val Acc: {val_acc:.4f}")

        if val_acc > best_val_acc:
            best_val_acc = val_acc
            torch.save(model.state_dict(), save_path)
            print(f"💾 Best model updated! Val Acc = {best_val_acc:.4f} → {save_path}")

    # ===== SIMPAN HISTORY =====
    history = {
        "train_loss": train_losses,
        "val_loss":   val_losses,
        "train_acc":  train_accs,
        "val_acc":    val_accs
    }
    with open(history_path, "wb") as f:
        pickle.dump(history, f)
    print("📊 History saved →", history_path)

    # ===== SIMPAN CSV =====
    df = pd.DataFrame({
        "epoch": list(range(1, len(train_losses)+1)),
        "train_loss": train_losses,
        "val_loss":   val_losses,
        "train_acc":  train_accs,
        "val_acc":    val_accs
    })
    df.to_csv(csv_path, index=False)
    print("📑 CSV saved →", csv_path)

    print(f"✅ Training selesai. Best Val Acc: {best_val_acc:.4f}")
    return history_path, csv_path
```

```python
def compute_map50(model, data_loader, device, num_classes,
                  score_thresh=0.3, iou_thresh=0.5):
    from torchvision.ops import box_iou
    import numpy as np
    model.eval()

    total_gt = 0
    total_tp = 0

    with torch.no_grad():
        for images, targets in data_loader:
            images = [img.to(device) for img in images]
            preds = model(images)

            for pred, tgt in zip(preds, targets):
                gt_boxes  = tgt["boxes"].cpu()
                gt_labels = tgt["labels"].cpu()
                total_gt += len(gt_labels)

                pb = pred["boxes"].cpu()
                pl = pred["labels"].cpu()
                ps = pred["scores"].cpu()

                keep = ps >= score_thresh
                pb, pl = pb[keep], pl[keep]

                if len(pb) == 0 or len(gt_boxes) == 0:
                    continue

                ious = box_iou(pb, gt_boxes)
                max_iou, idx = ious.max(dim=1)

                for i in range(len(max_iou)):
                    if max_iou[i] >= iou_thresh and pl[i] == gt_labels[idx[i]]:
                        total_tp += 1

    if total_gt == 0:
        return 0.0
    return total_tp / total_gt
```

```python
import torch.optim as optim
from torch import amp
from tqdm.auto import tqdm
import os, pickle
import pandas as pd

def train_ssd(model, train_loader, val_loader, device,
              num_classes_det,
              epochs=30,
              lr=1e-4,
              save_path="/content/models/ssd_mobilenet_best_retrain.pt",
              history_path="/content/models/ssd_mobilenet_history.pkl",
              csv_path="/content/models/ssd_mobilenet_results.csv"):

    os.makedirs(os.path.dirname(save_path), exist_ok=True)

    model.to(device)
    scaler = amp.GradScaler("cuda", enabled=(device.type == "cuda"))
    optimizer = optim.AdamW(
        [p for p in model.parameters() if p.requires_grad],
        lr=lr
    )

    train_losses, val_losses, map_values = [], [], []
    best_map = 0.0

    for epoch in range(epochs):
        print(f"\n==== EPOCH {epoch+1}/{epochs} ====")

        # ===== TRAIN =====
        model.train()
        running_loss = 0.0
        pbar = tqdm(train_loader, desc=f"[SSD] Train E{epoch+1}", leave=True)

        for images, targets in pbar:
            images  = [img.to(device) for img in images]
            targets = [{k: v.to(device) for k, v in t.items()} for t in targets]

            optimizer.zero_grad()

            with amp.autocast("cuda", enabled=(device.type=="cuda")):
                loss_dict = model(images, targets)

            # jaga-jaga kalau ada batch aneh
            if not isinstance(loss_dict, dict):
                continue

            loss = sum(loss_dict.values())

            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()

            running_loss += loss.item()
            pbar.set_postfix({"loss": f"{loss.item():.3f}"})

        train_loss = running_loss / max(1, len(train_loader))

        # ===== VAL LOSS =====
        total_val_loss = 0.0
        valid_val_batches = 0

        with torch.no_grad():
            model.train()
            for images, targets in val_loader:
                images  = [img.to(device) for img in images]
                targets = [{k: v.to(device) for k, v in t.items()} for t in targets]

                with amp.autocast("cuda", enabled=(device.type=="cuda")):
                    loss_dict = model(images, targets)

                if not isinstance(loss_dict, dict):
                    continue

                loss = sum(loss_dict.values())
                total_val_loss += loss.item()
                valid_val_batches += 1

        if valid_val_batches > 0:
            val_loss = total_val_loss / valid_val_batches
        else:
            val_loss = 0.0

        mAP = compute_map50(model, val_loader, device,
                            num_classes=num_classes_det,
                            score_thresh=0.3, iou_thresh=0.5)

        train_losses.append(train_loss)
        val_losses.append(val_loss)
        map_values.append(mAP)

        print(f"[SSD] Epoch {epoch+1}/{epochs} | "
              f"Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f} | "
              f"mAP@0.5: {mAP:.4f}")

        if mAP > best_map:
            best_map = mAP
            torch.save(model.state_dict(), save_path)
            print(f"💾 Best model updated → {save_path}")

    # SIMPAN HISTORY
    history = {
        "train_loss": train_losses,
        "val_loss": val_losses,
        "map": map_values
    }
    with open(history_path, "wb") as f:
        pickle.dump(history, f)

    print("📊 History saved:", history_path)

    # SIMPAN CSV
    df = pd.DataFrame({
        "epoch": list(range(1, len(train_losses)+1)),
        "train_loss": train_losses,
        "val_loss": val_losses,
        "mAP@0.5": map_values
    })
    df.to_csv(csv_path, index=False)
    print("📑 CSV saved:", csv_path)

    print(f"🎯 Retrain selesai — Best mAP@0.5: {best_map:.4f}")
    return history_path, csv_path
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
from tqdm import tqdm
import torchvision.ops as ops

class_names = ["Anorganik", "B3", "Logam", "Organik"]  #
num_classes = len(class_names)

def evaluate_confusion_matrix(model, loader, device, score_thresh=0.3, iou_thresh=0.5):
    model.eval()
    all_gt = []
    all_pred = []

    with torch.no_grad():
        for images, targets in tqdm(loader, desc="Evaluating CM"):
            images  = [img.to(device) for img in images]
            targets = [{k: v.to(device) for k, v in t.items()} for t in targets]

            outputs = model(images)

            for pred, tgt in zip(outputs, targets):
                gt_boxes  = tgt["boxes"]
                gt_labels = tgt["labels"]

                if gt_boxes.numel() == 0:
                    continue

                keep = pred["scores"] >= score_thresh
                pred_boxes  = pred["boxes"][keep]
                pred_labels = pred["labels"][keep]

                if pred_boxes.numel() == 0:
                    continue

                iou = ops.box_iou(pred_boxes, gt_boxes)
                max_iou, idx = iou.max(dim=1)

                pred_match = pred_labels[max_iou >= iou_thresh].cpu().numpy()
                gt_match   = gt_labels[idx[max_iou >= iou_thresh]].cpu().numpy()

                all_gt.extend(gt_match.tolist())
                all_pred.extend(pred_match.tolist())

    return np.array(all_gt), np.array(all_pred)


gt_arr, pred_arr = evaluate_confusion_matrix(ssd_model, val_det_loader, device)

print("\nUnique GT Labels: ", np.unique(gt_arr, return_counts=True))
print("Unique Pred Labels:", np.unique(pred_arr, return_counts=True))

cm = confusion_matrix(gt_arr, pred_arr, labels=list(range(num_classes)))

print("\nConfusion Matrix:\n", cm)

print("\nClassification Report:")
print(classification_report(gt_arr, pred_arr, target_names=class_names, zero_division=0))


plt.figure(figsize=(7,6))
sns.heatmap(cm, annot=True, fmt="d",
            xticklabels=class_names,
            yticklabels=class_names,
            cmap="Blues")
plt.title("SSD MobileNet - Confusion Matrix (IoU ≥ 0.5)")
plt.xlabel("Predicted")
plt.ylabel("Ground Truth")
plt.show()
```
