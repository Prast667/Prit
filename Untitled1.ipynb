```python
import os

# Folder utama tempat dataset diekstrak
EXTRACT_PATH = "/content/dataset"

# Nama folder dataset utama
DATASET_NAME = "Tikusv1"
DATASET_PATH = os.path.join(EXTRACT_PATH, DATASET_NAME)

# --- otomatis deteksi nama folder validasi ---
if os.path.exists(os.path.join(DATASET_PATH, "val")):
    VAL_NAME = "val"
elif os.path.exists(os.path.join(DATASET_PATH, "valid")):
    VAL_NAME = "valid"
else:
    raise FileNotFoundError("Folder validasi tidak ditemukan. Pastikan bernama 'val' atau 'valid'.")

# --- definisi path utama ---
train_dir = os.path.join(DATASET_PATH, "train")
val_dir   = os.path.join(DATASET_PATH, VAL_NAME)
test_dir  = os.path.join(DATASET_PATH, "test")

print("‚úÖ Path dataset terdeteksi otomatis:")
print(f"Train: {train_dir}")
print(f"Val  : {val_dir}")
print(f"Test : {test_dir}")

# --- validasi isi kelas ---
print("\nüìÇ Daftar kelas (train):")
print(os.listdir(train_dir))
```

```python
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# Augmentasi & normalisasi dataset
train_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(15),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])

val_test_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])

# Load dataset
train_data = datasets.ImageFolder(train_dir, transform=train_transform)
val_data   = datasets.ImageFolder(val_dir, transform=val_test_transform)
test_data  = datasets.ImageFolder(test_dir, transform=val_test_transform)

# DataLoader
BATCH_SIZE = 32
train_loader = DataLoader(train_data, batch_size=BATCH_SIZE, shuffle=True, num_workers=2, pin_memory=True)
val_loader   = DataLoader(val_data, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)
test_loader  = DataLoader(test_data, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)

# Informasi kelas
class_names = train_data.classes
num_classes = len(class_names)
print(f"‚úÖ Jumlah kelas: {num_classes}")
print("üß© Classes:", class_names)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report

# Deteksi GPU otomatis
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("üî• Device aktif:", device)

# Konfigurasi umum
EPOCHS = 40
LEARNING_RATE = 1e-3
PATIENCE = 5
torch.manual_seed(42)
```

```python
def train_model(model, name):
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)
    scaler = torch.amp.GradScaler(device="cuda")

    train_losses, val_losses = [], []
    train_accs, val_accs = [], []
    best_acc = 0.0

    for epoch in range(EPOCHS):
        model.train()
        running_loss, correct_train, total_train = 0.0, 0, 0

        # === Training ===
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()

            with torch.amp.autocast(device_type="cuda"):
                outputs = model(inputs)
                loss = criterion(outputs, labels)

            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()

            running_loss += loss.item() * inputs.size(0)
            _, preds = torch.max(outputs, 1)
            correct_train += (preds == labels).sum().item()
            total_train += labels.size(0)

        train_loss = running_loss / len(train_loader.dataset)
        train_acc = correct_train / total_train

        # === Validation ===
        model.eval()
        val_loss, correct_val, total_val = 0, 0, 0
        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                with torch.amp.autocast(device_type="cuda"):
                    outputs = model(inputs)
                    loss = criterion(outputs, labels)
                val_loss += loss.item() * inputs.size(0)
                _, preds = torch.max(outputs, 1)
                correct_val += (preds == labels).sum().item()
                total_val += labels.size(0)

        val_loss /= len(val_loader.dataset)
        val_acc = correct_val / total_val

        train_losses.append(train_loss)
        val_losses.append(val_loss)
        train_accs.append(train_acc)
        val_accs.append(val_acc)

        print(f"[{name}] Epoch {epoch+1}/{EPOCHS} | "
              f"Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f} | "
              f"Train Acc: {train_acc:.4f} | Val Acc: {val_acc:.4f}")

        if val_acc > best_acc:
            best_acc = val_acc
            torch.save(model.state_dict(), f"/content/drive/MyDrive/models/{name}_best.pt")

    # === Plot Curves ===
    plt.figure(figsize=(10,4))

    # Plot Loss
    plt.subplot(1,2,1)
    plt.plot(train_losses, label="Train Loss")
    plt.plot(val_losses, label="Val Loss")
    plt.title(f"Loss Curves - {name}")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()

    # Plot Accuracy
    plt.subplot(1,2,2)
    plt.plot(train_accs, label="Train Acc")
    plt.plot(val_accs, label="Val Acc")
    plt.title(f"Accuracy Curves - {name}")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend()

    plt.show()

    return best_acc
```

```python
cnn_model = nn.Sequential(
    nn.Conv2d(3, 32, 3, padding=1), nn.ReLU(), nn.MaxPool2d(2),
    nn.Conv2d(32, 64, 3, padding=1), nn.ReLU(), nn.MaxPool2d(2),
    nn.Conv2d(64, 128, 3, padding=1), nn.ReLU(), nn.MaxPool2d(2),
    nn.Flatten(),
    nn.Linear(128 * 28 * 28, 256), nn.ReLU(),
    nn.Linear(256, num_classes)
).to(device)

cnn_acc = train_model(cnn_model, "CNN")
```

```python
from torchvision import models

resnet_model = models.resnet50(weights="IMAGENET1K_V1")
for param in resnet_model.parameters():
    param.requires_grad = True

resnet_model.fc = nn.Linear(resnet_model.fc.in_features, num_classes)
resnet_model = resnet_model.to(device)

resnet_acc = train_model(resnet_model, "ResNet50")
```

```python
densenet_model = models.densenet121(weights="IMAGENET1K_V1")
for param in densenet_model.parameters():
    param.requires_grad = True

densenet_model.classifier = nn.Linear(densenet_model.classifier.in_features, num_classes)
densenet_model = densenet_model.to(device)

densenet_acc = train_model(densenet_model, "DenseNet121")
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
from IPython.display import display, Markdown # Tambahkan ini jika Anda menggunakan display/Markdown

# =========================================================================
# ‚ö†Ô∏è 1. DEFINISI VARIABEL GLOBAL (Contoh - Ganti dengan aktual Anda) ‚ö†Ô∏è
# =========================================================================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
class_names = ['Penggerek', 'Tikus', 'Ulat', 'Walang Sangit', 'Wereng', 'burung pipit', 'orong-orong']
# Variabel 'test_loader' dan 'model_obj' harus didefinisikan sebelum dipanggil.
# =========================================================================


def test_model(model_path, model_name, model_obj):
    """
    Menguji model klasifikasi dan membuat visualisasi Confusion Matrix tanpa garis grid.
    """

    # ‚ö†Ô∏è Cek dependensi utama (Hapus baris ini jika Anda yakin test_loader sudah didefinisikan)
    if 'test_loader' not in globals():
        raise NameError("Variabel 'test_loader' belum didefinisikan! Harap buat PyTorch DataLoader untuk data uji Anda.")

    # Load Weights Model
    model_obj.load_state_dict(torch.load(model_path))
    model_obj.eval()
    all_preds, all_labels = [], []

    # Inference Loop (sama seperti sebelumnya)
    for imgs, labels in test_loader:
        imgs, labels = imgs.to(device), labels.to(device)
        with torch.no_grad():
            outputs = model_obj(imgs)
            _, preds = torch.max(outputs, 1)

        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

    # Hitung Akurasi
    acc = np.mean(np.array(all_preds) == np.array(all_labels))
    print(f"‚úÖ {model_name} Test Accuracy: {acc:.4f}")

    # Classification Report
    print("\n" + "="*50)
    print(classification_report(all_labels, all_preds, target_names=class_names))
    print("="*50 + "\n")

    # Confusion Matrix Visualization
    cm = confusion_matrix(all_labels, all_preds)

    plt.figure(figsize=(8,6))

    # üí° PERUBAHAN UTAMA: MENGHILANGKAN GARIS GRID
    sns.heatmap(cm,
                annot=True,
                fmt="d",
                cmap="Blues",
                # Parameter ini diubah menjadi nol untuk menghilangkan garis
                linewidths=0.0,
                xticklabels=class_names,
                yticklabels=class_names)

    plt.title(f"Confusion Matrix - {model_name}", fontsize=14, fontweight='bold')
    plt.xlabel("Predicted", fontsize=12)
    plt.ylabel("True", fontsize=12)
    plt.show()

    return acc
```

```python
# ============================================================
# üì¶ PREDIKSI DAN VISUALISASI HASIL DARI 5 MODEL CNN
# ============================================================

import os
import random
import torch
import torch.nn as nn
import torchvision.models as models
import torchvision.transforms as transforms
from torchvision import datasets
from PIL import Image
import matplotlib.pyplot as plt

# ============================================================
# 1Ô∏è‚É£ SETUP DASAR
# ============================================================

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"‚úÖ Device aktif: {device}")

# Folder dataset test
test_dir = "/content/dataset/Tikusv1/test"

# Folder tempat model disimpan
model_dir = "/content/drive/MyDrive/models"

# Daftar kelas (ganti sesuai dataset kamu)
class_names = sorted(os.listdir(test_dir))
print("üìÅ Kelas yang terdeteksi:", class_names)

# ============================================================
# 2Ô∏è‚É£ FUNGSI UNTUK MEMBUAT MODEL DAN LOAD WEIGHT
# ============================================================

def create_model(model_name, num_classes):
    """Buat model sesuai nama dan jumlah kelas, langsung di-device"""
    if model_name == "CNN":
        model = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),
            nn.Flatten(),
            nn.Linear(128 * 28 * 28, 256),  # untuk input 224x224
            nn.ReLU(),
            nn.Linear(256, num_classes)
        )
        return model.to(device)

    elif model_name == "ResNet50":
        model = models.resnet50(weights=None)
        model.fc = nn.Linear(model.fc.in_features, num_classes)
        return model.to(device)

    elif model_name == "DenseNet121":
        model = models.densenet121(weights=None)
        model.classifier = nn.Linear(model.classifier.in_features, num_classes)
        return model.to(device)

    elif model_name == "MobileNetV2":
        model = models.mobilenet_v2(weights=None)
        model.classifier[1] = nn.Linear(model.classifier[1].in_features, num_classes)
        return model.to(device)

    elif model_name == "EfficientNetB0":
        model = models.efficientnet_b0(weights=None)
        model.classifier[1] = nn.Linear(model.classifier[1].in_features, num_classes)
        return model.to(device)

    else:
        raise ValueError(f"Model {model_name} tidak dikenal!")

def safe_load_model(model_name):
    """Load model dengan toleransi partial load"""
    model_paths = {
        "CNN": os.path.join(model_dir, "CNN_best.pt"),
        "ResNet50": os.path.join(model_dir, "ResNet50_best.pt"),
        "DenseNet121": os.path.join(model_dir, "DenseNet121_best.pt"),
        "MobileNetV2": os.path.join(model_dir, "MobileNetV2_best.pt"),
        "EfficientNetB0": os.path.join(model_dir, "EfficientNetB0_best.pt")
    }

    path = model_paths.get(model_name)
    if not os.path.exists(path):
        print(f"‚ö†Ô∏è File model {path} tidak ditemukan, dilewati.")
        return None

    model = create_model(model_name, len(class_names))
    state_dict = torch.load(path, map_location=device)

    # Load partial state_dict
    missing_keys, unexpected_keys = model.load_state_dict(state_dict, strict=False)
    if len(missing_keys) > 0 or len(unexpected_keys) > 0:
        print(f"‚ö†Ô∏è {model_name} loaded partially:")
        if len(missing_keys) > 0:
            print(f"   Missing keys: {missing_keys[:5]}{'...' if len(missing_keys) > 5 else ''}")
        if len(unexpected_keys) > 0:
            print(f"   Unexpected keys: {unexpected_keys[:5]}{'...' if len(unexpected_keys) > 5 else ''}")
    else:
        print(f"‚úÖ {model_name} loaded fully.")

    model.eval()
    print(f"‚úÖ Model {model_name} berhasil dimuat dari {path}")
    return model

# ============================================================
# 3Ô∏è‚É£ LOAD SEMUA MODEL
# ============================================================

cnn_model = safe_load_model("CNN")
resnet_model = safe_load_model("ResNet50")
densenet_model = safe_load_model("DenseNet121")
mobilenet_model = safe_load_model("MobileNetV2")
efficientnet_model = safe_load_model("EfficientNetB0")

models_dict = {
    "CNN": cnn_model,
    "ResNet50": resnet_model,
    "DenseNet121": densenet_model,
    "MobileNetV2": mobilenet_model,
    "EfficientNetB0": efficientnet_model
}

# ============================================================
# 4Ô∏è‚É£ PILIH GAMBAR UJI SECARA ACAK
# ============================================================

all_images = []
for root, dirs, files in os.walk(test_dir):
    for f in files:
        if f.lower().endswith((".jpg", ".png", ".jpeg")):
            all_images.append(os.path.join(root, f))

if len(all_images) == 0:
    raise FileNotFoundError(f"Tidak ada gambar ditemukan di {test_dir}")

test_image_path = random.choice(all_images)
print(f"üñºÔ∏è Gambar uji: {test_image_path}")

# ============================================================
# 5Ô∏è‚É£ FUNGSI PREDIKSI
# ============================================================

def predict_image(model, img_path):
    model.eval()
    image = Image.open(img_path).convert("RGB")
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406],
                             [0.229, 0.224, 0.225])
    ])
    image_tensor = transform(image).unsqueeze(0).to(device)
    with torch.no_grad():
        outputs = model(image_tensor)
        _, preds = torch.max(outputs, 1)
    return preds.item(), image

# ============================================================
# 6Ô∏è‚É£ VISUALISASI HASIL PREDIKSI
# ============================================================

fig, axes = plt.subplots(1, 5, figsize=(22, 6))
for i, (name, model) in enumerate(models_dict.items()):
    if model is not None:
        pred_idx, img = predict_image(model, test_image_path)
        axes[i].imshow(img)
        axes[i].set_title(f"{name}\nPrediksi: {class_names[pred_idx]}", fontsize=12)
        axes[i].axis("off")
    else:
        axes[i].set_visible(False)

plt.suptitle("Hasil Prediksi Lima Model CNN", fontsize=16)
plt.tight_layout()
plt.show()
```
