```python
import os
import glob
import yaml
from PIL import Image
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms

# ==== 1. Baca data.yaml ====
DATASET_PATH = "/content/Project-Automated-Kontolodon-10"
yaml_path = os.path.join(DATASET_PATH, "data.yaml")

with open(yaml_path, "r") as f:
    cfg = yaml.safe_load(f)

class_names = cfg["names"]   # ["organik", "anorganik", "logam", "b3"]
num_classes = cfg["nc"]      # 4

print("Classes from data.yaml:", class_names)
print("num_classes:", num_classes)

# ==== 2. Transform ====
train_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(15),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])

val_test_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])

# ==== 3. Dataset custom dari YOLO labels ====
class WasteDataset(Dataset):
    def __init__(self, img_dir, label_dir, transform=None):
        self.img_dir = img_dir
        self.label_dir = label_dir
        self.transform = transform

        # semua gambar di folder images
        self.img_paths = sorted(
            glob.glob(os.path.join(img_dir, "*.*"))
        )

    def __len__(self):
        return len(self.img_paths)

    def __getitem__(self, idx):
        img_path = self.img_paths[idx]

        # nama dasar file (tanpa ekstensi)
        base = os.path.splitext(os.path.basename(img_path))[0]

        # label dengan nama yang sama
        label_path = os.path.join(self.label_dir, base + ".txt")

        # baca gambar
        img = Image.open(img_path).convert("RGB")

        # baca label YOLO (ambil class id baris pertama)
        with open(label_path, "r") as f:
            line = f.readline().strip().split()
            cls_idx = int(line[0])   # kolom pertama = id kelas (0..3)

        if self.transform:
            img = self.transform(img)

        return img, cls_idx  # untuk CrossEntropyLoss

# ==== 4. Buat dataset & dataloader ====
train_img_dir = os.path.join(DATASET_PATH, "train", "images")
train_lbl_dir = os.path.join(DATASET_PATH, "train", "labels")

val_name = "valid" if os.path.exists(os.path.join(DATASET_PATH,"valid")) else "val"
val_img_dir = os.path.join(DATASET_PATH, val_name, "images")
val_lbl_dir = os.path.join(DATASET_PATH, val_name, "labels")

test_img_dir = os.path.join(DATASET_PATH, "test", "images")
test_lbl_dir = os.path.join(DATASET_PATH, "test", "labels")

train_data = WasteDataset(train_img_dir, train_lbl_dir, transform=train_transform)
val_data   = WasteDataset(val_img_dir,   val_lbl_dir,   transform=val_test_transform)
test_data  = WasteDataset(test_img_dir,  test_lbl_dir,  transform=val_test_transform)

BATCH_SIZE = 32
train_loader = DataLoader(train_data, batch_size=BATCH_SIZE, shuffle=True,  num_workers=2, pin_memory=True)
val_loader   = DataLoader(val_data,   batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)
test_loader  = DataLoader(test_data,  batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)

print("Train samples:", len(train_data))
print("Val samples  :", len(val_data))
print("Test samples :", len(test_data))
print("Classes:", class_names)
```

```python
import os
import glob
import shutil
import yaml

# ====== PATH YOLO & OUTPUT CLASSIFICATION ======
YOLO_PATH = "/content/Project-Automated-Kontolodon-10"
OUT_PATH  = "/content/waste_classification"  # dataset baru format classification

os.makedirs(OUT_PATH, exist_ok=True)

# ====== BACA data.yaml UNTUK NAMA KELAS ======
yaml_path = os.path.join(YOLO_PATH, "data.yaml")
with open(yaml_path, "r") as f:
    cfg = yaml.safe_load(f)

class_names = cfg["names"]   # contoh: ["organik","anorganik","logam","b3"]
print("Classes:", class_names)

# ====== FUNGSI BANTU CARI GAMBAR BERDASARKAN NAMA DASAR ======
def find_image(img_dir, base):
    exts = [".jpg", ".jpeg", ".png", ".bmp", ".JPG", ".PNG"]
    for ext in exts:
        path = os.path.join(img_dir, base + ext)
        if os.path.exists(path):
            return path
    return None

# ====== KONVERSI UNTUK SETIAP SPLIT ======
splits = ["train", "valid", "test"]  # ganti 'valid' ke 'val' kalau di kamu namanya val

for split in splits:
    src_img_dir = os.path.join(YOLO_PATH, split, "images")
    src_lbl_dir = os.path.join(YOLO_PATH, split, "labels")

    if not os.path.exists(src_img_dir):
        print(f"⚠️ Lewati {split}, folder images tidak ada:", src_img_dir)
        continue

    label_files = sorted(glob.glob(os.path.join(src_lbl_dir, "*.txt")))
    print(f"\n=== Split: {split} | {len(label_files)} label files ===")

    n_copied = 0
    n_empty  = 0
    n_missing_img = 0

    for label_path in label_files:
        base = os.path.splitext(os.path.basename(label_path))[0]

        # baca baris pertama yang berisi sesuatu
        cls_id = None
        with open(label_path, "r") as f:
            for line in f:
                line = line.strip()
                if not line:
                    continue
                tokens = line.split()
                cls_id = int(tokens[0])
                break

        # kalau file label kosong / tidak ada bbox -> skip
        if cls_id is None:
            n_empty += 1
            continue

        cls_name = class_names[cls_id]

        img_path = find_image(src_img_dir, base)
        if img_path is None:
            n_missing_img += 1
            continue

        dst_dir = os.path.join(OUT_PATH, split, cls_name)
        os.makedirs(dst_dir, exist_ok=True)

        dst_path = os.path.join(dst_dir, os.path.basename(img_path))
        shutil.copy2(img_path, dst_path)
        n_copied += 1

    print(f"📦 Copied: {n_copied}")
    print(f"⚠️ Label kosong (skip): {n_empty}")
    print(f"⚠️ Gambar tidak ditemukan (skip): {n_missing_img}")

print("\n✅ Konversi selesai. Struktur baru ada di:", OUT_PATH)
print("Isi /train:", os.listdir(os.path.join(OUT_PATH, "train")))
```

```python
import os
import torch
import torch.nn as nn
import torch.optim as optim
from torch.cuda.amp import GradScaler, autocast
import matplotlib.pyplot as plt

def train_model(
    model,
    name,
    train_loader,
    val_loader,
    device,
    epochs=EPOCHS,
    lr=LEARNING_RATE
):
    model = model.to(device)

    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)

    # AMP hanya aktif jika menggunakan CUDA
    use_amp = (device == "cuda" or device.type == "cuda")
    scaler = GradScaler(enabled=use_amp)

    train_losses, val_losses = [], []
    train_accs, val_accs = [], []
    best_acc = 0.0

    # ==== Simpan di penyimpanan lokal Colab ====
    save_dir = "/content/models"
    os.makedirs(save_dir, exist_ok=True)
    best_model_path = os.path.join(save_dir, f"{name}_best.pt")

    for epoch in range(epochs):
        # ===== Training =====
        model.train()
        running_loss, correct_train, total_train = 0.0, 0, 0

        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)

            optimizer.zero_grad()

            if use_amp:
                with autocast():
                    outputs = model(inputs)
                    loss = criterion(outputs, labels)
                scaler.scale(loss).backward()
                scaler.step(optimizer)
                scaler.update()
            else:
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()

            running_loss += loss.item() * inputs.size(0)
            _, preds = torch.max(outputs, 1)
            correct_train += (preds == labels).sum().item()
            total_train += labels.size(0)

        train_loss = running_loss / len(train_loader.dataset)
        train_acc = correct_train / total_train

        # ===== Validation =====
        model.eval()
        val_loss, correct_val, total_val = 0.0, 0, 0

        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)

                if use_amp:
                    with autocast():
                        outputs = model(inputs)
                        loss = criterion(outputs, labels)
                else:
                    outputs = model(inputs)
                    loss = criterion(outputs, labels)

                val_loss += loss.item() * inputs.size(0)
                _, preds = torch.max(outputs, 1)
                correct_val += (preds == labels).sum().item()
                total_val += labels.size(0)

        val_loss /= len(val_loader.dataset)
        val_acc = correct_val / total_val

        train_losses.append(train_loss)
        val_losses.append(val_loss)
        train_accs.append(train_acc)
        val_accs.append(val_acc)

        print(
            f"[{name}] Epoch {epoch+1}/{epochs} | "
            f"Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f} | "
            f"Train Acc: {train_acc:.4f} | Val Acc: {val_acc:.4f}"
        )

        # Simpan model terbaik
        if val_acc > best_acc:
            best_acc = val_acc
            torch.save(model.state_dict(), best_model_path)

    # ===== Plot Curves =====
    plt.figure(figsize=(10, 4))

    plt.subplot(1, 2, 1)
    plt.plot(train_losses, label="Train Loss")
    plt.plot(val_losses, label="Val Loss")
    plt.title(f"Loss Curves - {name}")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(train_accs, label="Train Acc")
    plt.plot(val_accs, label="Val Acc")
    plt.title(f"Accuracy Curves - {name}")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend()

    plt.tight_layout()
    plt.show()

    print(f"\n📌 Model terbaik disimpan di: {best_model_path}")
    return best_acc
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class ConvBNAct(nn.Module):
    def __init__(self, in_ch, out_ch, kernel_size=3, stride=1, groups=1):
        super().__init__()
        padding = kernel_size // 2
        self.conv = nn.Conv2d(in_ch, out_ch, kernel_size, stride,
                              padding, groups=groups, bias=False)
        self.bn = nn.BatchNorm2d(out_ch)
        self.act = nn.SiLU(inplace=True)

    def forward(self, x):
        return self.act(self.bn(self.conv(x)))

class DWSeparableResidualBlock(nn.Module):
    def __init__(self, in_ch, out_ch, stride=1):
        super().__init__()
        self.use_residual = (stride == 1 and in_ch == out_ch)

        self.dw = ConvBNAct(in_ch, in_ch, kernel_size=3,
                            stride=stride, groups=in_ch)
        self.pw = ConvBNAct(in_ch, out_ch, kernel_size=1, stride=1)

        if not self.use_residual:
            self.shortcut = nn.Sequential(
                nn.Conv2d(in_ch, out_ch, kernel_size=1, stride=stride, bias=False),
                nn.BatchNorm2d(out_ch)
            )
        else:
            self.shortcut = nn.Identity()

    def forward(self, x):
        out = self.pw(self.dw(x))
        return out + self.shortcut(x)

class WasteNet(nn.Module):
    def __init__(self, num_classes: int):
        super().__init__()
        self.stem = nn.Sequential(
            ConvBNAct(3, 32, kernel_size=3, stride=2),  # 224->112
            nn.MaxPool2d(3, stride=2, padding=1)        # 112->56
        )

        self.stage1 = nn.Sequential(
            DWSeparableResidualBlock(32, 32, stride=1),
            DWSeparableResidualBlock(32, 32, stride=1),
        )

        self.stage2 = nn.Sequential(
            DWSeparableResidualBlock(32, 64, stride=2),
            DWSeparableResidualBlock(64, 64, stride=1),
        )

        self.stage3 = nn.Sequential(
            DWSeparableResidualBlock(64, 128, stride=2),
            DWSeparableResidualBlock(128, 128, stride=1),
            DWSeparableResidualBlock(128, 128, stride=1),
        )

        self.stage4 = nn.Sequential(
            DWSeparableResidualBlock(128, 256, stride=2),
            DWSeparableResidualBlock(256, 256, stride=1),
            DWSeparableResidualBlock(256, 256, stride=1),
        )

        self.gap = nn.AdaptiveAvgPool2d((1, 1))

        self.classifier = nn.Sequential(
            nn.Flatten(),         # 256
            nn.Dropout(0.5),
            nn.Linear(256, 512),
            nn.SiLU(inplace=True),
            nn.Dropout(0.5),
            nn.Linear(512, num_classes)
        )

    def forward(self, x):
        x = self.stem(x)
        x = self.stage1(x)
        x = self.stage2(x)
        x = self.stage3(x)
        x = self.stage4(x)
        x = self.gap(x)
        x = self.classifier(x)
        return x
```

```python
import os
import torch.optim as optim
from torch import amp
import matplotlib.pyplot as plt

EPOCHS = 30
LEARNING_RATE = 1e-4

def train_model(model, name, train_loader, val_loader, device,
                epochs=EPOCHS, lr=LEARNING_RATE):

    model = model.to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)

    use_amp = (device.type == "cuda")
    scaler = amp.GradScaler("cuda", enabled=use_amp)

    train_losses, val_losses = [], []
    train_accs, val_accs = [], []
    best_acc = 0.0

    save_dir = "/content/models"
    os.makedirs(save_dir, exist_ok=True)
    best_model_path = os.path.join(save_dir, f"{name}_best.pt")

    for epoch in range(epochs):
        # ----- TRAIN -----
        model.train()
        running_loss, correct_train, total_train = 0.0, 0, 0

        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()

            with amp.autocast("cuda", enabled=use_amp):
                outputs = model(inputs)
                loss = criterion(outputs, labels)

            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()

            running_loss += loss.item() * inputs.size(0)
            _, preds = torch.max(outputs, 1)
            correct_train += (preds == labels).sum().item()
            total_train += labels.size(0)

        train_loss = running_loss / len(train_loader.dataset)
        train_acc = correct_train / total_train

        # ----- VAL -----
        model.eval()
        val_loss, correct_val, total_val = 0.0, 0, 0
        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                with amp.autocast("cuda", enabled=use_amp):
                    outputs = model(inputs)
                    loss = criterion(outputs, labels)

                val_loss += loss.item() * inputs.size(0)
                _, preds = torch.max(outputs, 1)
                correct_val += (preds == labels).sum().item()
                total_val += labels.size(0)

        val_loss /= len(val_loader.dataset)
        val_acc = correct_val / total_val

        train_losses.append(train_loss)
        val_losses.append(val_loss)
        train_accs.append(train_acc)
        val_accs.append(val_acc)

        print(f"[{name}] Epoch {epoch+1}/{epochs} | "
              f"Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f} | "
              f"Train Acc: {train_acc:.4f} | Val Acc: {val_acc:.4f}")

        if val_acc > best_acc:
            best_acc = val_acc
            torch.save(model.state_dict(), best_model_path)

    # plot kurva
    plt.figure(figsize=(10,4))
    plt.subplot(1,2,1)
    plt.plot(train_losses, label="Train Loss")
    plt.plot(val_losses, label="Val Loss")
    plt.title(f"Loss Curves - {name}")
    plt.xlabel("Epoch"); plt.ylabel("Loss"); plt.legend()

    plt.subplot(1,2,2)
    plt.plot(train_accs, label="Train Acc")
    plt.plot(val_accs, label="Val Acc")
    plt.title(f"Accuracy Curves - {name}")
    plt.xlabel("Epoch"); plt.ylabel("Accuracy"); plt.legend()

    plt.tight_layout()
    plt.show()

    print("📌 Best model saved at:", best_model_path)
    return best_acc
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
from IPython.display import display, Markdown # Tambahkan ini jika Anda menggunakan display/Markdown

# =========================================================================
# ⚠️ 1. DEFINISI VARIABEL GLOBAL (Contoh - Ganti dengan aktual Anda) ⚠️
# =========================================================================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
class_names = ['Penggerek', 'Tikus', 'Ulat', 'Walang Sangit', 'Wereng', 'burung pipit', 'orong-orong']
# Variabel 'test_loader' dan 'model_obj' harus didefinisikan sebelum dipanggil.
# =========================================================================


def test_model(model_path, model_name, model_obj):
    """
    Menguji model klasifikasi dan membuat visualisasi Confusion Matrix tanpa garis grid.
    """

    # ⚠️ Cek dependensi utama (Hapus baris ini jika Anda yakin test_loader sudah didefinisikan)
    if 'test_loader' not in globals():
        raise NameError("Variabel 'test_loader' belum didefinisikan! Harap buat PyTorch DataLoader untuk data uji Anda.")

    # Load Weights Model
    model_obj.load_state_dict(torch.load(model_path))
    model_obj.eval()
    all_preds, all_labels = [], []

    # Inference Loop (sama seperti sebelumnya)
    for imgs, labels in test_loader:
        imgs, labels = imgs.to(device), labels.to(device)
        with torch.no_grad():
            outputs = model_obj(imgs)
            _, preds = torch.max(outputs, 1)

        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

    # Hitung Akurasi
    acc = np.mean(np.array(all_preds) == np.array(all_labels))
    print(f"✅ {model_name} Test Accuracy: {acc:.4f}")

    # Classification Report
    print("\n" + "="*50)
    print(classification_report(all_labels, all_preds, target_names=class_names))
    print("="*50 + "\n")

    # Confusion Matrix Visualization
    cm = confusion_matrix(all_labels, all_preds)

    plt.figure(figsize=(8,6))

    # 💡 PERUBAHAN UTAMA: MENGHILANGKAN GARIS GRID
    sns.heatmap(cm,
                annot=True,
                fmt="d",
                cmap="Blues",
                # Parameter ini diubah menjadi nol untuk menghilangkan garis
                linewidths=0.0,
                xticklabels=class_names,
                yticklabels=class_names)

    plt.title(f"Confusion Matrix - {model_name}", fontsize=14, fontweight='bold')
    plt.xlabel("Predicted", fontsize=12)
    plt.ylabel("True", fontsize=12)
    plt.show()

    return acc
```

```python
import os
import random
import torch
import torch.nn as nn
import torchvision.models as models
import torchvision.transforms as transforms
from torchvision import datasets
from PIL import Image
import matplotlib.pyplot as plt
import timm
import json
import numpy as np
import pandas as pd
from sklearn.metrics import classification_report, accuracy_score

# ============================================================
# 0️⃣ SETUP DAN MUAT KELAS
# ============================================================

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
test_dir = "/content/dataset/Tikusv1/test"
model_dir = "/content/drive/MyDrive/models"

try:
    class_names = sorted(os.listdir(test_dir))
except FileNotFoundError:
    class_names = ['Penggerek', 'Tikus', 'Ulat', 'Walang Sangit', 'Wereng', 'burung pipit', 'orong-orong']

NUM_CLASSES = len(class_names)
print(f"✅ Device aktif: {device}")
print("📁 Kelas yang terdeteksi:", class_names)

# ============================================================
# 1️⃣ FUNGSI UTAMA (create_model, safe_load_model, test_model)
# ============================================================

def create_model(model_name, num_classes):
    """Buat model sesuai nama dan jumlah kelas."""
    if model_name == "CNN":
        model = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, padding=1), nn.ReLU(), nn.MaxPool2d(2, 2),
            nn.Conv2d(32, 64, kernel_size=3, padding=1), nn.ReLU(), nn.MaxPool2d(2, 2),
            nn.Conv2d(64, 128, kernel_size=3, padding=1), nn.ReLU(), nn.MaxPool2d(2, 2),
            nn.Flatten(),
            nn.Linear(128 * 28 * 28, 256),  nn.ReLU(),
            nn.Linear(256, num_classes)
        )
    elif model_name == "ResNet50":
        model = models.resnet50(weights=None)
        model.fc = nn.Linear(model.fc.in_features, num_classes)
    elif model_name == "DenseNet121":
        model = models.densenet121(weights=None)
        model.classifier = nn.Linear(model.classifier.in_features, num_classes)
    elif model_name == "MobileNetV2":
        model = timm.create_model("mobilenetv2_100", pretrained=False)
        model.classifier = nn.Sequential(nn.Dropout(0.2), nn.Linear(model.classifier.in_features, num_classes))
    elif model_name == "EfficientNetB0":
        model = timm.create_model("efficientnet_b0", pretrained=False)
        model.classifier = nn.Linear(model.classifier.in_features, num_classes)
    else:
        raise ValueError(f"Model {model_name} tidak dikenal!")
    return model.to(device)

def safe_load_model(model_name):
    """Load model untuk inisialisasi awal dengan perbaikan UnpicklingError."""
    path = os.path.join(model_dir, f"{model_name}_best.pt")
    if not os.path.exists(path):
        print(f"⚠️ File model {path} tidak ditemukan, dilewati.")
        return None, []

    model = create_model(model_name, NUM_CLASSES)

    # PERBAIKAN UNPICKLINGERROR
    try:
        state_dict = torch.load(path, map_location=device, weights_only=False)
    except Exception as e:
        print(f"❌ Gagal memuat state_dict untuk {model_name} dari {path}: {e}")
        return None, []

    # PERBAIKAN KLASIFIKATOR MOBILENETV2 (seperti di fungsi test_model)
    if model_name == "MobileNetV2":
        if 'classifier.weight' in state_dict and 'classifier.1.weight' not in state_dict:
            state_dict['classifier.1.weight'] = state_dict.pop('classifier.weight')
            state_dict['classifier.1.bias'] = state_dict.pop('classifier.bias')
            print("🔧 Perbaikan classifier MobileNetV2 berhasil diterapkan saat inisialisasi.")

    missing_keys, unexpected_keys = model.load_state_dict(state_dict, strict=False)

    is_fully_loaded = not missing_keys and not unexpected_keys
    if not is_fully_loaded:
        print(f"⚠️ {model_name} loaded partially:")
        print(f"   Missing keys: {list(missing_keys)[:5]}")
    else:
        print(f"✅ {model_name} loaded fully.")

    model.eval()
    return model, list(missing_keys)


def test_model(model_path, model_name, model_instance):
    """
    Fungsi pengujian model.
    Menggunakan model_instance yang sudah diinisialisasi untuk akurasi.

    Catatan: Fungsi ini tetap memuat bobot dari path untuk memastikan model yang diuji
    adalah yang terbaik.
    """

    # 1. Load Data Test
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])
    test_dataset = datasets.ImageFolder(test_dir, transform=transform)
    test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=32, shuffle=False)

    # 2. Load Weights ke model_instance
    try:
        # Load state dict dengan penanganan UnpicklingError
        state_dict = torch.load(model_path, map_location=device, weights_only=False)
    except Exception as e:
        print(f"❌ Gagal memuat state_dict untuk {model_name} dari {model_path}: {e}")
        return 0.0 # Kembali 0 jika gagal load

    # 👇 PERBAIKAN UTAMA MOBILE NETV2 di test_model
    if model_name == "MobileNetV2":
        if 'classifier.weight' in state_dict and 'classifier.1.weight' not in state_dict:
            state_dict['classifier.1.weight'] = state_dict.pop('classifier.weight')
            state_dict['classifier.1.bias'] = state_dict.pop('classifier.bias')

    # Lakukan load bobot
    missing_keys, unexpected_keys = model_instance.load_state_dict(state_dict, strict=False)

    if missing_keys or unexpected_keys:
        print(f"⚠️ {model_name} loaded partially in test_model:")
        if missing_keys: print(f"   Missing keys: {list(missing_keys)}")
        if unexpected_keys: print(f"   Unexpected keys: {list(unexpected_keys)}")

    model_instance.eval()

    all_preds = []
    all_targets = []

    # 3. Pengujian
    with torch.no_grad():
        for inputs, targets in test_loader:
            inputs, targets = inputs.to(device), targets.to(device)
            outputs = model_instance(inputs)
            _, preds = torch.max(outputs, 1)

            all_preds.extend(preds.cpu().numpy())
            all_targets.extend(targets.cpu().numpy())

    # 4. Hasil
    test_accuracy = accuracy_score(all_targets, all_preds)
    print(f"\n✅ {model_name} Test Accuracy: {test_accuracy:.4f}")

    # Laporan Klasifikasi
    report = classification_report(all_targets, all_preds, target_names=class_names, zero_division=0)
    print("\n==================================================")
    print(report)
    print("==================================================")

    return test_accuracy

# ============================================================
# 2️⃣ LOAD MODEL UNTUK INSTANCE DAN PENGUJIAN
# ============================================================

cnn_model, cnn_keys = safe_load_model("CNN")
resnet_model, resnet_keys = safe_load_model("ResNet50")
densenet_model, densenet_keys = safe_load_model("DenseNet121")
mobilenet_model, mobile_keys = safe_load_model("MobileNetV2")
efficientnet_model, efficient_keys = safe_load_model("EfficientNetB0")

# ============================================================
# 3️⃣ JALANKAN PENGUJIAN DAN HASILKAN CSV
# ============================================================

results = []

# Catatan: Path MobileNetV2 telah diperbaiki dari kesalahan 'MyDrive/MyDrive'
if cnn_model is not None:
    results.append({"Model": "CNN", "Test_Accuracy": test_model(os.path.join(model_dir, "CNN_best.pt"), "CNN", cnn_model)})
if resnet_model is not None:
    # ResNet50 mungkin masih gagal jika file .pt rusak
    if resnet_model is not None:
         results.append({"Model": "ResNet50", "Test_Accuracy": test_model(os.path.join(model_dir, "ResNet50_best.pt"), "ResNet50", resnet_model)})
    else:
         results.append({"Model": "ResNet50", "Test_Accuracy": 0.0}) # Jika gagal load awal

if densenet_model is not None:
    results.append({"Model": "DenseNet121", "Test_Accuracy": test_model(os.path.join(model_dir, "DenseNet121_best.pt"), "DenseNet121", densenet_model)})

# MobileNetV2 sekarang menggunakan path yang benar dan logika loading yang benar
if mobilenet_model is not None:
    results.append({"Model": "MobileNetV2", "Test_Accuracy": test_model(os.path.join(model_dir, "MobileNetV2_best.pt"), "MobileNetV2", mobilenet_model)})
else:
    results.append({"Model": "MobileNetV2", "Test_Accuracy": 0.0}) # Jika gagal load awal

if efficientnet_model is not None:
    results.append({"Model": "EfficientNetB0", "Test_Accuracy": test_model(os.path.join(model_dir, "EfficientNetB0_best.pt"), "EfficientNetB0", efficientnet_model)})
else:
    results.append({"Model": "EfficientNetB0", "Test_Accuracy": 0.0}) # Jika gagal load awal


df_results = pd.DataFrame(results)
df_results.to_csv("/content/drive/MyDrive/results_comparison.csv", index=False)
print("\n📊 Final Results:\n", df_results)

# ============================================================
# 4️⃣ VISUALISASI PREDIKSI GAMBAR UJI (Sesuai Permintaan Sebelumnya)
# ============================================================

# --- FUNGSI PREDIKSI ---
def predict_image(model, img_path):
    model.eval()
    image = Image.open(img_path).convert("RGB")
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])
    image_tensor = transform(image).unsqueeze(0).to(device)
    with torch.no_grad():
        outputs = model(image_tensor)
        _, preds = torch.max(outputs, 1)
    return preds.item(), image

# --- PILIH GAMBAR UJI SPESIFIK ---
target_classes = ['Ulat', 'orong-orong', 'Wereng', 'Tikus']
selected_test_images = []

for target_class in target_classes:
    class_path = os.path.join(test_dir, target_class)
    if os.path.exists(class_path):
        class_images = [os.path.join(class_path, f) for f in os.listdir(class_path) if f.lower().endswith((".jpg", ".png", ".jpeg"))]
        if class_images:
            selected_test_images.append(random.choice(class_images))

# --- VISUALISASI ---
models_dict = {
    "CNN": cnn_model, "ResNet50": resnet_model, "DenseNet121": densenet_model,
    "MobileNetV2": mobilenet_model, "EfficientNetB0": efficientnet_model
}
# Asumsi missing_keys_dict dibuat di luar sini jika diperlukan untuk label (dihapus untuk skrip lengkap)

num_images = len(selected_test_images)
num_loaded_models = sum(1 for model in models_dict.values() if model is not None)

if num_loaded_models > 0:
    fig, axes = plt.subplots(num_images, num_loaded_models, figsize=(4 * num_loaded_models, 5 * num_images))

    # Menangani array 1D atau 0D
    if num_images == 1 and num_loaded_models > 1: axes = [axes]
    elif num_images > 1 and num_loaded_models == 1: axes = [[ax] for ax in axes]
    elif num_images == 1 and num_loaded_models == 1: axes = [[axes]]

    for img_idx, image_path in enumerate(selected_test_images):
        plot_idx = 0
        true_class = image_path.split(os.sep)[-2]

        for name, model in models_dict.items():
            if model is not None:
                pred_idx, img = predict_image(model, image_path)
                ax = axes[img_idx][plot_idx]
                ax.imshow(img)

                if img_idx == 0: ax.set_title(f"{name}", fontsize=12)
                if plot_idx == 0: ax.set_ylabel(f"Asli: {true_class}", fontsize=12, rotation=90, labelpad=15)

                predicted_class = class_names[pred_idx]
                color = 'green' if predicted_class == true_class else 'red'

                ax.set_xlabel(f"Prediksi: {predicted_class}", color=color, fontsize=10)
                ax.set_xticks([]); ax.set_yticks([])

                plot_idx += 1

    plt.suptitle(f"Hasil Prediksi ({num_loaded_models} Model CNN)", fontsize=16, y=1.02)
    plt.tight_layout()
    plt.show()
```

```python
import time

def train_model(model, name):
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)
    scaler = torch.amp.GradScaler(device="cuda")

    train_losses, val_losses = [], []
    train_accs, val_accs = [], []
    best_acc = 0.0
    total_train_time = 0 # Initialize total training time

    for epoch in range(EPOCHS):
        start_time = time.time() # Start time for the epoch

        model.train()
        running_loss, correct_train, total_train = 0.0, 0, 0

        # === Training ===
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()

            with torch.amp.autocast(device_type="cuda"):
                outputs = model(inputs)
                loss = criterion(outputs, labels)

            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()

            running_loss += loss.item() * inputs.size(0)
            _, preds = torch.max(outputs, 1)
            correct_train += (preds == labels).sum().item()
            total_train += labels.size(0)

        train_loss = running_loss / len(train_loader.dataset)
        train_acc = correct_train / total_train

        # === Validation ===
        model.eval()
        val_loss, correct_val, total_val = 0, 0, 0
        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                with torch.amp.autocast(device_type="cuda"):
                    outputs = model(inputs)
                    loss = criterion(outputs, labels)
                val_loss += loss.item() * inputs.size(0)
                _, preds = torch.max(outputs, 1)
                correct_val += (preds == labels).sum().item()
                total_val += labels.size(0)

        val_loss /= len(val_loader.dataset)
        val_acc = correct_val / total_val

        end_time = time.time() # End time for the epoch
        epoch_time = end_time - start_time
        total_train_time += epoch_time # Accumulate total training time

        train_losses.append(train_loss)
        val_losses.append(val_loss)
        train_accs.append(train_acc)
        val_accs.append(val_acc)

        print(f"[{name}] Epoch {epoch+1}/{EPOCHS} | "
              f"Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f} | "
              f"Train Acc: {train_acc:.4f} | Val Acc: {val_acc:.4f} | "
              f"Epoch Time: {epoch_time:.2f}s") # Print epoch time

        if val_acc > best_acc:
            best_acc = val_acc
            torch.save(model.state_dict(), f"/content/drive/MyDrive/models/{name}_best.pt")

    # === Plot Curves ===
    plt.figure(figsize=(10,4))

    # Plot Loss
    plt.subplot(1,2,1)
    plt.plot(train_losses, label="Train Loss")
    plt.plot(val_losses, label="Val Loss")
    plt.title(f"Loss Curves - {name}")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()

    # Plot Accuracy
    plt.subplot(1,2,2)
    plt.plot(train_accs, label="Train Acc")
    plt.plot(val_accs, label="Val Acc")
    plt.title(f"Accuracy Curves - {name}")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend()

    plt.show()

    total_train_time_hours = total_train_time / 3600 # Convert seconds to hours
    print(f"[{name}] Total Training Time: {total_train_time_hours:.2f} hours")

    return best_acc, total_train_time_hours # Return best accuracy and total training time
```
